```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
%matplotlib inline
```

```python
iris = sn.load_dataset('iris')
#统计条形图
sn.countplot(x='species',data=iris)
```

```python
#直方图（sepal_length）
sn.distplot(iris['petal_length'],color="green")
```

```python
#条形图
sn.barplot(x='species',y='petal_length',data=iris)
```

```python
#箱型图
sn.boxplot(x='species',y='petal_length',data=iris)
```

```python
#散点图
sn.regplot(x='petal_length',y='petal_width',data=iris,color='violet')
```

```python
#分类绘图
g = sn.FacetGrid(iris,hue='species')
g.map(sn.distplot,'petal_length').add_legend()
```

```python
#分类线性回归
r = sn.FacetGrid(iris,hue='species')
r.set(xlim=(0,2.5))
r.map(sn.regplot,'petal_width','petal_length').add_legend()
```
